### Feature Understanding

```python
df['Year_Introduced'].value_counts()
```

### Feature Relationships

```python
## Checking Correlation
cor = df[['Year_Introduced','Speed_mph',
                 'Height_ft','Inversions','Gforce']].dropna().corr()
cor
```

### Asking Question about the data

```python
df.head()
```

```python
### What are the Location with the fastest roller coasters(minimum of 10 coasters at that location)?
df['Location'].value_counts()
```

```python
loc = df[df['Location']!='Other'].groupby(
    ['Location'])['Speed_mph'].agg(['mean','count'])
loc
```

```python
top_speed = loc[loc['count']>=10].sort_values('mean')
top_speed
```

```python
plt.figure(figsize=(14,8))
sns.barplot(y=top_speed.index,x=top_speed['mean'])
plt.title('Average Speed by Locations')
plt.xlabel('Average Speed')
plt.ylabel('Locations')
plt.show()
```
