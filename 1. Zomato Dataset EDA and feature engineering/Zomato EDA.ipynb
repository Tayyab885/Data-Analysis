### Read Country code dataset

```python
df_country = pd.read_excel("Dataset/Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

#### As we can see both datasets have Country Code so we can concat them together on the base of that.

```python
final_df =pd.merge(df,df_country,on='Country Code',how='left')
# "on" tells on which column you want to join and "how means" which dataset to piroritize (df or df_column)
```

```python
final_df.head()
```

```python
## To check datatype
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

### Let's draw some graphs on the base on countries

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
## Draw Pie Char
plt.pie(country_values,labels=country_names)
```

### **See It looks very bad because of so many countries**

```python
## Draw Pie Char for top 3 countries
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

So from that we can write the Observation like Zomato have about 94% customers from India and 5% from America and 1% from UK.

### Let's analyse the dataset more

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text'])
```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### Observation
1. When rating is between 4.5-4.9---->Excellent
2. When rating is between 4.0-4.4---->Very Good
3. When rating is between 3.5-3.9---->Good
4. When rating is between 2.5-3.4---->Average
5. When rating is between 1.8-2.4---->Poor

### Let's draw diagram

```python
ratings.head()
```

```python
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

### Let's color according to the color format in ratings

```python
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue = "Rating color")
```

### Se the Colors are not matching so to do that

```python
matplotlib.rcParams['figure.figsize'] = (12,6)
# since white is not visible we use blue
sns.barplot(x="Aggregate rating",y="Rating Count",hue = "Rating color",data=ratings,palette=["blue","red","orange","yellow","green","green"])
```

### Observations
1. Most of the users didn't rate
2. Max number of ratings are between2.5 to 3.4

```python
# Count plot
# Count plot is used for categorical Values
sns.countplot(x="Rating color",data = ratings, palette=["blue","red","orange","yellow","green","green"])
```
